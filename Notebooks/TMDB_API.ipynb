# API call to create a dataframe that contain all the movie recommendation

```python
# Testing api call
import requests

url = "https://api.themoviedb.org/3/movie/862/recommendations?language=en-US&page=1"

headers = {
    "accept": accept,
    "Authorization": Authorization
}

response = requests.get(url, headers=headers).json()

print(response['results'])
```

```python
# API call for the first 50 movies in the links_df
movie_titles = []
movie_id = []
movie_plot = []
movie_poster = []
movie_genre = []
movie_imdbid_list = []

# Get recommendations for the first 50 movies from the api 
for i in range(0,50):
    print(i)
    movie_tmdbid = links_df["tmdbId"][i]
    
    
    #Set up url 
    url = "https://api.themoviedb.org/3/movie/" + str(movie_tmdbid) + "/recommendations?language=en-US&page=1"
    
    #Set up headers -- .env contains keys 
    headers = {
        "accept": accept,
        "Authorization": Authorization
    }
        
    response = requests.get(url, headers=headers).json()
    for j in range(0,10):
        movie_titles.append(response['results'][j]['title'])
        movie_id.append(response['results'][j]['id'])
        movie_plot.append(response['results'][j]['overview'])
        movie_genre.append(response['results'][j]['genre_ids'])
```

```python
# Adding the first 50 movies in links_df into the dataframe
for i in range(0,50):
    print(i)
    movie_tmdbid = links_df["tmdbId"][i]
    
    #Set up url 
    url = "https://api.themoviedb.org/3/movie/" + str(movie_tmdbid) + "?language=en-US"
    

    headers = {
        "accept": accept,
        "Authorization": Authorization
    }

    response = requests.get(url, headers=headers).json()
#     print(response)
    movie_titles.append(response['title'])
    movie_id.append(response['id'])
    movie_plot.append(response['overview'])
    movie_genre.append(response['genres'][0]['id'])
```

```python
imdb_id = []
for i in movie_id:
    print(i)
    #Set up url 
    url = "https://api.themoviedb.org/3/movie/" + str(i) + "?language=en-US"


    headers = {
        "accept": accept,
        "Authorization": Authorization
    }

    response = requests.get(url, headers=headers).json()
    imdb_id.append(response["imdb_id"])
```

# Rewrite poster path from omdbapi 

```python
# API url parameters
url = "http://www.omdbapi.com/?i="
movie_id = "tt0114709"
key = "apikey=" + API_KEY + "&"
```

```python
# Loop through each title and get poster path 
poster_list = []

for i in range(0,len(test)):
    print(i)
    
    movie_id = test["imdb_id"][i]
    url1 = url + str(movie_id) + "&" + key
    response = requests.get(url1).json()
    poster_list.append(response["Poster"])
```

```python
test["Poster"] = poster_list
```
