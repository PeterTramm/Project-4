```python
# Compute the cosine similarity matrix
cosine_sim = chi2_kernel(song_prediction.loc[song_prediction["clusters"]==7], song_prediction.loc[song_prediction["clusters"]==7])
```

```python
# Function that takes in song title as input and outputs most similar songs
def get_recommendations(title):
    # Get the index of the movie that matches the title
    cluster_number = song_clusters.loc[song_clusters["track_name"]==title]['Clusters']
    
    a = song_clusters.loc[song_clusters["Clusters"]==int(cluster_number)].reset_index()
    
    b = a.drop(columns=["artist(s)_name","artist_count","released_year","released_month","released_day",
                        "in_spotify_playlists",'in_spotify_charts', 'streams', 'in_apple_playlists', 
                        'in_apple_charts','in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts']).set_index("track_name")
    c = b.copy()
    c = c.drop(columns=["key","mode","index"])
    
    dummies = pd.get_dummies(b[["mode","key"]])
    
    d= pd.concat([c, dummies],axis=1)
    
    cosine_sim = chi2_kernel(d, d)
    indices = pd.Series(a.index, index=a["track_name"]).drop_duplicates()
    idx = indices[title]

    
    # Get the pairwsie similarity scores of all songs with that song
    sim_scores = list(enumerate(cosine_sim[idx]))
 
    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
   
    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[0:11]
    
    sim_scores_list = [item[1] for item in sim_scores]
    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]
   
    
    recommendations = a[["track_name","artist(s)_name","Clusters"]].iloc[movie_indices]
    recommendations["sim_score"] = sim_scores_list
    # Return the top 10 most similar
    return recommendations
```
